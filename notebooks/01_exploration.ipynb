```python
import pandas as pd 
import numpy as np
import kaggle 
from dotenv import load_dotenv

import os

import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
# Load dotenv
load_dotenv()

# Import environment variables
KAGGLE_TOKEN_NAME = os.environ['KAGGLE_TOKEN_NAME']
KAGGLE_TOKE_VALUE = os.environ['KAGGLE_TOKEN_VALUE']
```

```python
# Download data if data is not available
if not os.path.isdir('../data/Ai_generated_dataset') or not os.path.isdir('../data/real_dataset'):
    # Authenticate credentials
    kaggle.api.authenticate()

    # Download data
    kaggle.api.dataset_download_files(dataset='rhythmghai/ai-vs-real-images-dataset', path='../data', unzip=True)
```

```python
# Import training dataset
dataset = image_dataset_from_directory(
    '../data',
    labels='inferred',
    image_size=(512, 512)
)

print(f'Class names: {dataset.class_names}')
```

```python
# Checkclass distribution
label_counts = {}
class_names = dataset.class_names

for images, labels in dataset:
    for label in labels.numpy():
        class_name = class_names[label]
        label_counts[class_name] = label_counts.get(class_name, 0) + 1

print(label_counts)

# Visualize class distribution
```
